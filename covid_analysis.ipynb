```python
import pandas as pd

# Load the dataset
df = pd.read_csv("data/owid-covid-data.csv")

# Preview the first 5 rows
df.head()
```

```python
# Display column names
print("ðŸ”‘ Columns in dataset:")
print(df.columns.tolist())

# Check basic info
df.info()

# Summary stats
df.describe()
```

```python
# Count missing values per column
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values[missing_values > 0]
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
selected_countries = ['Kenya', 'United States', 'India']
df_filtered = df[df['location'].isin(selected_countries)].copy()
```

```python
df_filtered.dropna(subset=['date', 'total_cases'], inplace=True)
```

```python
df_filtered.sort_values(by=['location', 'date'], inplace=True)
df_filtered.fillna(method='ffill', inplace=True)
```

```python
print("Remaining missing values after cleaning:")
print(df_filtered.isnull().sum().sort_values(ascending=False).head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_filtered, x='date', y='total_cases', hue='location')
plt.title('ðŸ“ˆ Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_filtered, x='date', y='total_deaths', hue='location')
plt.title('ðŸ’€ Total COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```
